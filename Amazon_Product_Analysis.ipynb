## **What is EDA?**

1. Exploratory Data Analysis (EDA) is a crucial process in data science and analytics.
2. It involves examining and visualizing data sets to understand their main characteristics.
3. EDA is typically the first step in the data analysis process.
4. It allows analysts to:
    
  *    Summarize the data.
  *    Uncover patterns.
  *    Identify anomalies.
  *    Test hypotheses.
  *    Check assumptions.

5. EDA uses statistical graphics and other data visualization techniques to achieve these goals.

## **Why do we require EDA?**

* Understanding Data Structure
* Identifying Data Quality Issues
* Uncovering Patterns and Relationships
* Hypothesis Testing
* Informing Model Selection
* Reducing the Risk of Errors
* Data Visualization and Communication


## **Import Libraries:**

```python
# Begin by importing the necessary libraries: NumPy, Pandas, Statistics, Matplotlib, Seaborn, etc.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Load Data:**

```python
## Read the CSV file into a Pandas DataFrame.

df = pd.read_csv('amazon.csv')
```

```python
df.head()
```

## **Data Cleaning:**

```python
# Check for missing values and handle them appropriately (e.g., dropping rows/columns, imputation).

df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
## Explore the data types of each column and ensure they are correct.

df.info()
```

```python
df.head().T
```

```python
df['actual_price'] = df['actual_price'].replace('[₹,]', '', regex=True)   ## regex -- regular expression
```

```python
df['discounted_price'] = df['discounted_price'].replace('[₹,]', '', regex=True)
```

```python
df.head()
```

```python
## Changing datatype of columns

df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['discount_percentage'] = df['discount_percentage'].replace('%', '', regex=True)
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df['rating']=pd.to_numeric(df['rating'],errors='coerce')
df['rating']
```

```python
df['rating_count'] = df['rating_count'].replace(',', '', regex=True)
```

```python
df['rating_count'] = df['rating_count'].astype(float)
```

```python
df.head()
```

## Question 2

```python
## What is the average rating across different product categories?
```

```python
df["df_category"] = df['category'].str.split('|',expand=True)[0]
```

```python
df['category'].str.split('|',expand=True)[0].value_counts()
```

```python
df_avg_rating1 = df.groupby('df_category')['rating'].mean().reset_index()
```

```python
sns.lineplot(df_avg_rating1,x='df_category',y='rating',color='purple',marker='*',markersize=10)
plt.xticks(rotation=75)
plt.title('Average rating across different product categories')
plt.show()
```

## Question 5

```python
## Are certain product categories more frequently discounted?
```

```python

# Group by category and calculate the discount frequency
discount_frequency = df.groupby('category')['discount_percentage'].count().reset_index()

# Rename columns for better understanding
discount_frequency.columns = ['category', 'discount_frequency']
```

```python
sorted_by_dis_freq = discount_frequency.sort_values(by='discount_frequency',ascending=False).head(15)
```

```python
sns.barplot(sorted_by_dis_freq,x='category',y='discount_frequency')
plt.title('product categories more frequently discounted')
plt.xticks(rotation=90)
plt.show()
```
